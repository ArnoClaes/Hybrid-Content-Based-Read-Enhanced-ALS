# Load data

```
drive.mount('/content/drive', force_remount=True)
```

```
df = pd.read_csv('/content/drive/My Drive/Knab/Data/CleanData/clean_page_data.csv')
df_url = pd.read_csv("/content/drive/My Drive/Knab/Data/CleanData/clean_article_data.csv") 
```

```
df_use = df[['URL', 'clientid_hashed', 'visitid', 'BiebYN']]
df_shift = df[['clientid_hashed', 'visitid', 'BiebYN']].rename(columns={'clientid_hashed': 'clientid_hashed_right', 'visitid': 'visitid_right', 'BiebYN':'BiebYN_right'})
```

```
df_left = df_use.drop(df_use.tail(1).index).reset_index(drop=True)
df_right = df_shift.drop(df_shift.head(1).index).reset_index(drop=True)
```

```
df_calculator = pd.concat([df_left,df_right], axis=1)
```

```
df_calculator = df_calculator[(df_calculator['clientid_hashed'] == df_calculator['clientid_hashed_right']) & (df_calculator['visitid'] == df_calculator['visitid_right']) & (df_calculator['BiebYN'] == 1) & (df_calculator['BiebYN_right'] == 0)]
```

```
df_calculator = df_calculator.assign(ArticleYN=df_calculator.URL.isin(df_url.URL).astype(int))
```

```
article_conv = df_calculator['ArticleYN'].sum()/len(df)
```

```
df_calculator[df_calculator['ArticleYN'] == 1]['URL'].value_counts()/df_calculator['ArticleYN'].sum()*100
```

```
print(f'Conversion from Article to Commercial is {article_conv * 100:.2f}% of total article clicks')
```

```
print(f'Conversion from Bieb to Commercial is {len(df_calculator)/len(df) * 100:.2f}% of total Bieb clicks')
```
